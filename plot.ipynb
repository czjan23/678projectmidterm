```
import matplotlib.pyplot as plt
import numpy as np

x = [0, 1, 2, 3, 4, 5, 6, 7]
file_size = ["1KB", "128KB", "256KB", "512KB", "1MB", "128MB", "256MB", "512MB"]

MA_ws = [89.25, 8242.77, 14192.73, 22065.84, 31578.84, 104391.91, 96246.50, 94636.50]
AWS_ws = [570.34, 48525.97, 86767.82, 142221.96, 206290.47, 104956.20, 83016.06, 72412.33]
GCP_ws = [225.46, 19343.65, 25077.60, 24989.17, 24128.69, 18565.70, 18653.27, 18685.80]

plt.plot(x, MA_ws, linestyle='-', marker='o', color='b', label='Microsoft Azure')
plt.plot(x, AWS_ws, linestyle='-', marker='v', color='y', label='AWS EC2')
plt.plot(x, GCP_ws, linestyle='-', marker='s', color='r', label='GCP')

plt.xlabel('File Size')
plt.ylabel('Writing Speed \(bytes/ms)')

plt.title("Mean Writing Speed of 3 Cloud Platform")
plt.xticks(x, file_size)
plt.legend()

plt.show()
```
